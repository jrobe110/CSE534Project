```python
from fabrictestbed_extensions.fablib.fablib import FablibManager as fablib_manager
fablib = fablib_manager()
```

Create slice with nodes

```python
slice = fablib.new_slice(name="project")

site1='UTAH'
site2='STAR'
site3='FIU'

print (f'The selected sites are {site1}, {site2}, {site3}') 

romeo = slice.add_node(name="romeo", 
                      site=site1, 
                      cores=4, 
                      ram=8, 
                      disk=20, 
                      image='default_ubuntu_20')

RouterA = slice.add_node(name="RouterA", 
                      site=site3, 
                      cores=4, 
                      ram=8, 
                      disk=20, 
                      image='default_ubuntu_20')

RouterB = slice.add_node(name="RouterB", 
                      site=site3, 
                      cores=4, 
                      ram=8, 
                      disk=20, 
                      image='default_ubuntu_20')

juliet = slice.add_node(name="juliet", 
                      site=site3, 
                      cores=4, 
                      ram=8, 
                      disk=20, 
                      image='default_ubuntu_20')

switch = slice.add_node(name="switch", 
                      site=site2, 
                      cores=32, 
                      ram=16, 
                      disk=40, 
                      image='default_ubuntu_20')
```

```python
romeo_iface = romeo.add_component(model='NIC_Basic').get_interfaces()[0]
RouterA_iface = RouterA.add_component(model='NIC_Basic').get_interfaces()[0]
RouterB_iface = RouterB.add_component(model='NIC_Basic').get_interfaces()[0]

RouterB_iface2 = RouterB.add_component(model='NIC_Basic', name='net3_iface').get_interfaces()[0]
RouterA_iface2 = RouterA.add_component(model='NIC_Basic', name='net2_iface').get_interfaces()[0]

juliet_iface = juliet.add_component(model='NIC_Basic').get_interfaces()[0]
juliet_iface2 = juliet.add_component(model='NIC_Basic', name='net2_iface2').get_interfaces()[0]

switch_iface1 = switch.add_component(model='NIC_Basic', name='net1_nic').get_interfaces()[0]
switch_iface2 = switch.add_component(model='NIC_Basic', name='net2_nic').get_interfaces()[0]
switch_iface3 = switch.add_component(model='NIC_Basic', name='net3_nic').get_interfaces()[0]

juliet_iface= juliet.add_component(model='NIC_Basic', name='net_juliet').get_interfaces()[0]
net_juliet = slice.add_l2network(name='net_juliet', type='L2Bridge', interfaces=[RouterB_iface2, juliet_iface])

juliet_iface2= juliet.add_component(model='NIC_Basic', name='net2_juliet').get_interfaces()[0]
net2_juliet = slice.add_l2network(name='net2_juliet', type='L2Bridge', interfaces=[RouterA_iface2, juliet_iface2])
```

```python
net1 = slice.add_l2network(name='net1', interfaces=[romeo_iface, switch_iface1])
net2 = slice.add_l2network(name='net2', interfaces=[switch_iface2, RouterA_iface])
net3 = slice.add_l2network(name='net3', interfaces=[switch_iface3, RouterB_iface])
```

```python
#Submit Sli ce Request
slice.submit();
```

```python
romeo_iface_MAC = '00:00:00:00:00:01'
switch_iface1_MAC = '00:00:00:00:00:02'
switch_iface2_MAC = '00:00:00:00:00:03'
RouterA_iface_MAC = '00:00:00:00:00:04'
switch_iface3_MAC = '00:00:00:00:00:05'
RouterB_iface_MAC = '00:00:00:00:00:06'
RouterB_iface2_MAC = '00:00:00:00:00:07'
juliet_iface_MAC = '00:00:00:00:00:08'
RouterA_iface2_MAC = '00:00:00:00:00:09'
juliet_ifac2_MAC = '00:00:00:00:00:10'
```

```python
#Creating the connections between each component
romeo_switch_subnet = "192.168.1.0/24"
romeo_ip = '192.168.1.10/24'
switch_ip1 = '192.168.1.1/24'

stdout, stderr = romeo.execute(f'sudo ifconfig {romeo_iface_name} {romeo_ip}')
stdout, stderr = switch.execute(f'sudo ifconfig {switch_iface1_name} {switch_ip1}')

stdout, stderr = romeo.execute(f'sudo ifconfig {romeo_iface_name} hw ether {romeo_iface_MAC}')
stdout, stderr = switch.execute(f'sudo ifconfig {switch_iface1_name} hw ether {switch_iface1_MAC}')

RouterA_switch_subnet = "192.168.2.0/24"
RouterA_ip = '192.168.2.10/24'
switch_ip2 = '192.168.2.1/24'

stdout, stderr = RouterA.execute(f'sudo ifconfig {RouterA_iface_name} {RouterA_ip}')
stdout, stderr = switch.execute(f'sudo ifconfig {switch_iface2_name} {switch_ip2}')

stdout, stderr = RouterA.execute(f'sudo ifconfig {RouterA_iface_name} hw ether {RouterA_iface_MAC}')
stdout, stderr = switch.execute(f'sudo ifconfig {switch_iface2_name} hw ether {switch_iface2_MAC}')

RouterB_switch_subnet = "192.168.3.0/24"
RouterB_ip = '192.168.3.10/24'
switch_ip3 = '192.168.3.1/24'

stdout, stderr = RouterB.execute(f'sudo ifconfig {RouterB_iface_name} {RouterB_ip}')
stdout, stderr = switch.execute(f'sudo ifconfig {switch_iface3_name} {switch_ip3}')

stdout, stderr = RouterB.execute(f'sudo ifconfig {RouterB_iface_name} hw ether {RouterB_iface_MAC}')
stdout, stderr = switch.execute(f'sudo ifconfig {switch_iface3_name} hw ether {switch_iface3_MAC}')

RouterB_juliet_subnet = "192.168.4.0/24"
RouterB_ip2 = '192.168.4.1/24'
juliet_ip = '192.168.4.10/24'

stdout, stderr = RouterB.execute(f'sudo ifconfig {RouterB_iface2_name} {RouterB_ip2}')
stdout, stderr = juliet.execute(f'sudo ifconfig {juliet_iface_name} {juliet_ip}')

stdout, stderr = RouterB.execute(f'sudo ifconfig {RouterB_iface2_name} hw ether {RouterB_iface2_MAC}')
stdout, stderr = juliet.execute(f'sudo ifconfig {juliet_iface_name} hw ether {juliet_iface_MAC}')


RouterA_juliet_subnet = "192.168.5.0/24"
RouterA_ip2 = '192.168.5.1/24'
juliet_ip2 = '192.168.5.10/24'

stdout, stderr = RouterA.execute(f'sudo ifconfig {RouterA_iface2_name} {RouterA_ip2}')
stdout, stderr = juliet.execute(f'sudo ifconfig {juliet_iface2_name} {juliet_ip2}')

stdout, stderr = RouterA.execute(f'sudo ifconfig {RouterA_iface2_name} hw ether {RouterA_iface2_MAC}')
stdout, stderr = juliet.execute(f'sudo ifconfig {juliet_iface2_name} hw ether {juliet_ifac2_MAC}')
```
